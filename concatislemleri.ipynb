```python
# ============================================================
# PANEL PROBIT ANALİZİ 
# ============================================================

import warnings
warnings.filterwarnings("ignore")

import os
import numpy as np
import pandas as pd
import statsmodels.api as sm

from statsmodels.discrete.discrete_model import Probit
from statsmodels.stats.outliers_influence import variance_inflation_factor

import matplotlib
matplotlib.use("Agg")
import matplotlib.pyplot as plt

import openpyxl


# ============================================================
# 1) AYARLAR
# ============================================================

DATA_PATH = r"C:\Users\dilara\Desktop\Hane halkı bilişim tüik\ferthane\bütünveriler.csv"
OUT_DIR = r"C:\Users\dilara\Desktop\Hane halkı bilişim tüik"

EXCEL_PATH = os.path.join(OUT_DIR, "panel_probit_sorunsuz_final.xlsx")
GRAFIK_PATH = os.path.join(OUT_DIR, "panel_probit_sorunsuz_grafik.png")


# ============================================================
# 2) VERİ YÜKLE
# ============================================================

COLS = [
    "referans yıl",
    "online sağlık hizmeti kullanımı",
    "eticaret ilaç alışverişi",
    "cinsiyet",
    "yaş",
    "eğitim",
    "hane gelir aylık",
    "hanenin aylık toplam net geliri (20'li gelir grubu)",
    "hanenin aylık toplam net geliri (5'li gelir grubu)",
]

print("=" * 70)
print("VERİ YÜKLENİYOR...")
print("=" * 70)

df_raw = pd.read_csv(DATA_PATH, usecols=COLS, low_memory=False)

print(f"Ham veri boyutu: {df_raw.shape}")


# ============================================================
# 3) TEMEL DEĞİŞKENLER
# ============================================================

df_raw["yil"] = pd.to_numeric(df_raw["referans yıl"], errors="coerce")
df_raw["yas"] = pd.to_numeric(df_raw["yaş"], errors="coerce")

df_raw["saglik_online"] = np.where(
    df_raw["online sağlık hizmeti kullanımı"] == 1,
    1,
    np.where(df_raw["online sağlık hizmeti kullanımı"] == 2, 0, np.nan)
)

df_raw["ilac_eticaret"] = np.where(
    df_raw["eticaret ilaç alışverişi"] == 1,
    1,
    np.where(df_raw["eticaret ilaç alışverişi"] == 2, 0, np.nan)
)


# ============================================================
# 4) EĞİTİM
# ============================================================

EGT_MAP = {
    0.0: 0,
    1.0: 1,
    2.0: 2,
    3.0: 3,
    4.0: 4,
    5.0: 5,
    51.0: 5,
    511.0: 5,
    512.0: 5,
    52.0: 5,
    53.0: 6,
    6.0: 6,
    7.0: 6,
}

df_raw["egitim_ord"] = pd.to_numeric(
    df_raw["eğitim"],
    errors="coerce"
).map(EGT_MAP)


# ============================================================
# 5) GELİRİ TEK 5'Lİ STANDARDA ÇEVİR
# ============================================================

df_raw["gelir_5li_std"] = np.nan

df_raw["gelir_kaynak"] = pd.Series(
    pd.NA,
    index=df_raw.index,
    dtype="object"
)

df_raw["gelir_aylik_num"] = pd.to_numeric(
    df_raw["hane gelir aylık"],
    errors="coerce"
)

for year in sorted(df_raw["yil"].dropna().unique()):

    mask_year = (
        (df_raw["yil"] == year) &
        df_raw["gelir_aylik_num"].notna() &
        (df_raw["gelir_aylik_num"] > 0)
    )

    if mask_year.sum() >= 5:

        try:
            gelir_5li = pd.qcut(
                df_raw.loc[mask_year, "gelir_aylik_num"],
                q=5,
                labels=[1, 2, 3, 4, 5],
                duplicates="drop"
            )

            df_raw.loc[mask_year, "gelir_5li_std"] = pd.to_numeric(
                gelir_5li,
                errors="coerce"
            )

            df_raw.loc[mask_year, "gelir_kaynak"] = "ayliktan_5liye"

        except Exception as e:
            print(f"{year} gelir dönüşüm hatası:", e)


def convert_20_to_5(x):

    if pd.isna(x):
        return np.nan

    try:
        x = int(float(x))
    except:
        return np.nan

    if 1 <= x <= 4:
        return 1
    elif 5 <= x <= 8:
        return 2
    elif 9 <= x <= 12:
        return 3
    elif 13 <= x <= 16:
        return 4
    elif 17 <= x <= 20:
        return 5

    return np.nan


gelir_20li = pd.to_numeric(
    df_raw["hanenin aylık toplam net geliri (20'li gelir grubu)"],
    errors="coerce"
)

mask_20li = (
    df_raw["gelir_5li_std"].isna() &
    gelir_20li.notna()
)

df_raw.loc[mask_20li, "gelir_5li_std"] = gelir_20li[mask_20li].apply(
    convert_20_to_5
)

df_raw.loc[mask_20li, "gelir_kaynak"] = "20liden_5liye"


gelir_5li_orj = pd.to_numeric(
    df_raw["hanenin aylık toplam net geliri (5'li gelir grubu)"],
    errors="coerce"
)

mask_5li = (
    df_raw["gelir_5li_std"].isna() &
    gelir_5li_orj.notna()
)

df_raw.loc[mask_5li, "gelir_5li_std"] = gelir_5li_orj[mask_5li]
df_raw.loc[mask_5li, "gelir_kaynak"] = "orijinal_5li"


print("\nGELİR KAYNAKLARI")
print(df_raw["gelir_kaynak"].value_counts(dropna=False))

print("\nGELİR 5'Lİ DAĞILIM")
print(df_raw["gelir_5li_std"].value_counts(dropna=False).sort_index())


# ============================================================
# 6) BAĞIMSIZ DEĞİŞKENLER
# ============================================================

df_raw["kadin"] = np.where(
    df_raw["cinsiyet"] == 2,
    1,
    np.where(df_raw["cinsiyet"] == 1, 0, np.nan)
)

df_raw["yas31_45"] = np.where(
    (df_raw["yas"] >= 31) & (df_raw["yas"] <= 45),
    1,
    0
)

df_raw["yas46_60"] = np.where(
    (df_raw["yas"] >= 46) & (df_raw["yas"] <= 60),
    1,
    0
)

df_raw["yas60ustu"] = np.where(
    df_raw["yas"] > 60,
    1,
    0
)

df_raw["egitim_orta_lise"] = np.where(
    df_raw["egitim_ord"].isin([3, 4]),
    1,
    0
)

df_raw["egitim_universite"] = np.where(
    df_raw["egitim_ord"] == 5,
    1,
    0
)

df_raw["egitim_uni_ustu"] = np.where(
    df_raw["egitim_ord"] == 6,
    1,
    0
)

df_raw["log_gelir"] = np.log(df_raw["gelir_5li_std"])


# ============================================================
# 7) YIL DUMMYLERİ
# Referans yıl: 2015
# ============================================================

for year in [
    2016, 2017, 2018, 2019, 2020,
    2021, 2022, 2023, 2024, 2025
]:

    df_raw[f"yil_{year}"] = np.where(
        df_raw["yil"] == year,
        1,
        0
    )


X_COLS = [
    "kadin",
    "log_gelir",
    "yas31_45",
    "yas46_60",
    "yas60ustu",
    "egitim_orta_lise",
    "egitim_universite",
    "egitim_uni_ustu",
    "yil_2016",
    "yil_2017",
    "yil_2018",
    "yil_2019",
    "yil_2020",
    "yil_2021",
    "yil_2022",
    "yil_2023",
    "yil_2024",
    "yil_2025"
]


# ============================================================
# 8) YARDIMCI FONKSİYONLAR
# ============================================================

def clean_X(X):

    X = X.copy()

    # sonsuz değerleri temizle
    X = X.replace([np.inf, -np.inf], np.nan)

    # tamamen boş sütunları sil
    X = X.dropna(axis=1, how="all")

    # tek değerli sütunları sil
    nunique = X.nunique(dropna=True)
    drop_cols = nunique[nunique <= 1].index.tolist()

    if len(drop_cols) > 0:
        print("\nModelden çıkarılan sabit değişkenler:")
        print(drop_cols)

    X = X.drop(columns=drop_cols, errors="ignore")

    return X


def safe_summary_table(model):

    try:
        return pd.DataFrame(model.summary2().tables[1])
    except Exception:
        return pd.DataFrame({
            "Coef.": model.params,
            "Std.Err.": model.bse,
            "z": model.tvalues,
            "P>|z|": model.pvalues,
            "[0.025": model.conf_int()[0],
            "0.975]": model.conf_int()[1]
        })


def safe_marginal_effects(model, X):

    try:
        margeff = model.get_margeff(at="overall")
        arr = margeff.margeff
        se = margeff.margeff_se
        z = margeff.tvalues
        p = margeff.pvalues
        ci = margeff.conf_int()

        names = [c for c in X.columns if c != "const"]

        min_len = min(
            len(names),
            len(arr),
            len(se),
            len(z),
            len(p),
            len(ci)
        )

        return pd.DataFrame({
            "Degisken": names[:min_len],
            "dy_dx": arr[:min_len],
            "Std_Hata": se[:min_len],
            "z": z[:min_len],
            "P_degeri": p[:min_len],
            "CI_alt": ci[:min_len, 0],
            "CI_ust": ci[:min_len, 1]
        })

    except Exception as e:
        return pd.DataFrame({
            "Hata": [str(e)]
        })


def run_model(dep_var, title):

    print("\n" + "=" * 70)
    print(title)
    print("=" * 70)

    X = df_raw[X_COLS].copy()
    y = df_raw[dep_var].copy()

    model_df = pd.concat([X, y], axis=1)
    model_df = model_df.replace([np.inf, -np.inf], np.nan)
    model_df = model_df.dropna()

    y = model_df[dep_var]
    X = model_df.drop(columns=[dep_var])

    X = clean_X(X)

    X = sm.add_constant(X, has_constant="add")

    print("Gözlem Sayısı:", len(model_df))
    print("Bağımlı değişken dağılımı:")
    print(y.value_counts())

    try:
        model = Probit(y, X).fit(
            method="newton",
            maxiter=500,
            disp=False
        )
    except Exception as e:
        print("Newton hata verdi, BFGS deneniyor:")
        print(e)

        model = Probit(y, X).fit(
            method="bfgs",
            maxiter=1000,
            disp=False
        )

    print(model.summary())

    summary_df = safe_summary_table(model)
    marginal_df = safe_marginal_effects(model, X)

    return model, summary_df, marginal_df, X, y


def calculate_vif(X):

    vif_X = X.copy()

    if "const" in vif_X.columns:
        vif_X = vif_X.drop(columns=["const"])

    vif_X = vif_X.astype(float)

    vif_df = pd.DataFrame()
    vif_df["Degisken"] = vif_X.columns

    vif_df["VIF"] = [
        variance_inflation_factor(vif_X.values, i)
        for i in range(vif_X.shape[1])
    ]

    return vif_df


# ============================================================
# 9) MODELLER
# ============================================================

model_online, online_summary, online_marginal, X_online, y_online = run_model(
    "saglik_online",
    "MODEL 1 - ONLINE SAĞLIK HİZMETİ"
)

model_eticaret, eticaret_summary, eticaret_marginal, X_eticaret, y_eticaret = run_model(
    "ilac_eticaret",
    "MODEL 2 - E-TİCARET İLAÇ ALIŞVERİŞİ"
)


# ============================================================
# 10) VIF
# ============================================================

vif_online = calculate_vif(X_online)
vif_eticaret = calculate_vif(X_eticaret)

print("\nVIF ONLINE")
print(vif_online)

print("\nVIF ETICARET")
print(vif_eticaret)


# ============================================================
# 11) GRAFİK
# ============================================================

fig, axes = plt.subplots(1, 2, figsize=(14, 5), sharey=True)

MODELS = [
    (model_online, y_online, "Online Sağlık", axes[0]),
    (model_eticaret, y_eticaret, "E-Ticaret İlaç", axes[1])
]

for model, y, title, ax in MODELS:

    probs = model.predict()

    plot_df = pd.DataFrame({
        "yil": df_raw.loc[y.index, "yil"],
        "prob": probs,
        "y": y
    })

    agg = plot_df.groupby("yil").agg(
        tahmin=("prob", "mean"),
        gozlem=("y", "mean")
    )

    x_pos = range(len(agg))

    ax.bar(
        x_pos,
        agg["tahmin"],
        alpha=0.7,
        label="Tahmin"
    )

    ax.plot(
        list(x_pos),
        agg["gozlem"],
        "o-",
        linewidth=2,
        label="Gözlenen"
    )

    ax.set_xticks(list(x_pos))
    ax.set_xticklabels(
        agg.index.astype(int),
        rotation=45
    )

    ax.set_title(title)
    ax.set_ylim(0, 1)
    ax.legend()

plt.tight_layout()
plt.savefig(GRAFIK_PATH, dpi=150, bbox_inches="tight")

print("\nGrafik kaydedildi:")
print(GRAFIK_PATH)


# ============================================================
# 12) EXCEL'E AKTAR
# ============================================================

with pd.ExcelWriter(EXCEL_PATH, engine="openpyxl") as writer:

    online_summary.to_excel(
        writer,
        sheet_name="Online_Probit"
    )

    online_marginal.to_excel(
        writer,
        sheet_name="Online_Marjinal",
        index=False
    )

    eticaret_summary.to_excel(
        writer,
        sheet_name="Eticaret_Probit"
    )

    eticaret_marginal.to_excel(
        writer,
        sheet_name="Eticaret_Marjinal",
        index=False
    )

    vif_online.to_excel(
        writer,
        sheet_name="VIF_Online",
        index=False
    )

    vif_eticaret.to_excel(
        writer,
        sheet_name="VIF_Eticaret",
        index=False
    )

    df_raw["gelir_kaynak"].value_counts(dropna=False).reset_index().to_excel(
        writer,
        sheet_name="Gelir_Kaynagi",
        index=False
    )

    df_raw["gelir_5li_std"].value_counts(dropna=False).sort_index().reset_index().to_excel(
        writer,
        sheet_name="Gelir_Dagilim",
        index=False
    )

    df_raw["yil"].value_counts(dropna=False).sort_index().reset_index().to_excel(
        writer,
        sheet_name="Yillara_Gore_Gozlem",
        index=False
    )

    pd.DataFrame({
        "Degisken": X_online.columns,
        "Modelde_Kullanildi": "Evet"
    }).to_excel(
        writer,
        sheet_name="Model_Degiskenleri",
        index=False
    )

    pd.DataFrame({
        "Bilgi": [
            "Toplam Satır",
            "Toplam Sütun",
            "Excel Yolu",
            "Grafik Yolu",
            "Not"
        ],
        "Deger": [
            df_raw.shape[0],
            df_raw.shape[1],
            EXCEL_PATH,
            GRAFIK_PATH,
            "Sabit/tek değerli değişkenler otomatik silindi."
        ]
    }).to_excel(
        writer,
        sheet_name="Genel_Bilgi",
        index=False
    )


print("\n" + "=" * 70)
print("ANALİZ TAMAMLANDI")
print("=" * 70)

print("\nExcel:")
print(EXCEL_PATH)

print("\nGrafik:")
print(GRAFIK_PATH)
```

```python
# ============================================================
# PANEL PROBIT ANALİZİ - FINAL KOD
# GELİR: 5'Lİ KATEGORİ
# REFERANS GELİR = GELİR 1 (EN DÜŞÜK)
# MODELE: gelir2, gelir3, gelir4, gelir5 girer
# ============================================================

import warnings
warnings.filterwarnings("ignore")

import os
import numpy as np
import pandas as pd
import statsmodels.api as sm

from statsmodels.discrete.discrete_model import Probit
from statsmodels.stats.outliers_influence import variance_inflation_factor

import matplotlib
matplotlib.use("Agg")
import matplotlib.pyplot as plt

import openpyxl


# ============================================================
# 1) AYARLAR
# ============================================================

DATA_PATH = r"C:\Users\dilara\Desktop\Hane halkı bilişim tüik\ferthane\bütünveriler.csv"

OUT_DIR = r"C:\Users\dilara\Desktop\Hane halkı bilişim tüik"

EXCEL_PATH = os.path.join(
    OUT_DIR,
    "panel_probit_gelir_dummy_final.xlsx"
)

GRAFIK_PATH = os.path.join(
    OUT_DIR,
    "panel_probit_gelir_dummy_grafik.png"
)


# ============================================================
# 2) VERİ YÜKLE
# ============================================================

COLS = [
    "referans yıl",
    "online sağlık hizmeti kullanımı",
    "eticaret ilaç alışverişi",
    "cinsiyet",
    "yaş",
    "eğitim",
    "hane gelir aylık",
    "hanenin aylık toplam net geliri (20'li gelir grubu)",
    "hanenin aylık toplam net geliri (5'li gelir grubu)",
]

print("=" * 70)
print("VERİ YÜKLENİYOR...")
print("=" * 70)

df_raw = pd.read_csv(
    DATA_PATH,
    usecols=COLS,
    low_memory=False
)

print("Ham veri boyutu:", df_raw.shape)


# ============================================================
# 3) TEMEL DEĞİŞKENLER
# ============================================================

df_raw["yil"] = pd.to_numeric(
    df_raw["referans yıl"],
    errors="coerce"
)

df_raw["yas"] = pd.to_numeric(
    df_raw["yaş"],
    errors="coerce"
)

# ------------------------------------------------------------
# Online sağlık
# ------------------------------------------------------------

df_raw["saglik_online"] = np.where(
    df_raw["online sağlık hizmeti kullanımı"] == 1,
    1,
    np.where(
        df_raw["online sağlık hizmeti kullanımı"] == 2,
        0,
        np.nan
    )
)

# ------------------------------------------------------------
# E-ticaret ilaç
# ------------------------------------------------------------

df_raw["ilac_eticaret"] = np.where(
    df_raw["eticaret ilaç alışverişi"] == 1,
    1,
    np.where(
        df_raw["eticaret ilaç alışverişi"] == 2,
        0,
        np.nan
    )
)


# ============================================================
# 4) EĞİTİM KODLAMA
# REFERANS: İLKOKUL VE ALTI
# ============================================================

egitim_num = pd.to_numeric(
    df_raw["eğitim"],
    errors="coerce"
)

# Ortaokul/Lise
df_raw["egitim_orta_lise"] = np.where(
    egitim_num.isin([3, 4]),
    1,
    np.where(egitim_num.notna(), 0, np.nan)
)

# Üniversite
df_raw["egitim_universite"] = np.where(
    egitim_num.isin([5, 51, 511, 512, 52]),
    1,
    np.where(egitim_num.notna(), 0, np.nan)
)

# Üniversite üstü
df_raw["egitim_uni_ustu"] = np.where(
    egitim_num.isin([53, 6, 7]),
    1,
    np.where(egitim_num.notna(), 0, np.nan)
)


# ============================================================
# 5) GELİRİ TEK 5'Lİ STANDART HALE GETİR
# ============================================================

df_raw["gelir_5li_std"] = np.nan

df_raw["gelir_kaynak"] = pd.Series(
    pd.NA,
    index=df_raw.index,
    dtype="object"
)

# ============================================================
# 5.1 AYLIK GELİR → HER YIL KENDİ İÇİNDE 5'Lİ GRUP
# ============================================================

df_raw["gelir_aylik_num"] = pd.to_numeric(
    df_raw["hane gelir aylık"],
    errors="coerce"
)

for year in sorted(df_raw["yil"].dropna().unique()):

    mask_year = (
        (df_raw["yil"] == year) &
        df_raw["gelir_aylik_num"].notna() &
        (df_raw["gelir_aylik_num"] > 0)
    )

    if mask_year.sum() >= 5:

        try:

            gelir_5li = pd.qcut(
                df_raw.loc[mask_year, "gelir_aylik_num"],
                q=5,
                labels=[1, 2, 3, 4, 5],
                duplicates="drop"
            )

            df_raw.loc[
                mask_year,
                "gelir_5li_std"
            ] = pd.to_numeric(
                gelir_5li,
                errors="coerce"
            )

            df_raw.loc[
                mask_year,
                "gelir_kaynak"
            ] = "ayliktan_5liye"

        except Exception as e:

            print(f"{year} gelir dönüşüm hatası:")
            print(e)


# ============================================================
# 5.2 20'Lİ GELİR → 5'Lİ GELİR
# ============================================================

def convert_20_to_5(x):

    if pd.isna(x):
        return np.nan

    try:
        x = int(float(x))
    except:
        return np.nan

    if 1 <= x <= 4:
        return 1

    elif 5 <= x <= 8:
        return 2

    elif 9 <= x <= 12:
        return 3

    elif 13 <= x <= 16:
        return 4

    elif 17 <= x <= 20:
        return 5

    return np.nan


gelir_20li = pd.to_numeric(
    df_raw["hanenin aylık toplam net geliri (20'li gelir grubu)"],
    errors="coerce"
)

mask_20li = (
    df_raw["gelir_5li_std"].isna() &
    gelir_20li.notna()
)

df_raw.loc[
    mask_20li,
    "gelir_5li_std"
] = gelir_20li[mask_20li].apply(convert_20_to_5)

df_raw.loc[
    mask_20li,
    "gelir_kaynak"
] = "20liden_5liye"


# ============================================================
# 5.3 HAZIR 5'Lİ GELİR
# ============================================================

gelir_5li_orj = pd.to_numeric(
    df_raw["hanenin aylık toplam net geliri (5'li gelir grubu)"],
    errors="coerce"
)

mask_5li = (
    df_raw["gelir_5li_std"].isna() &
    gelir_5li_orj.notna()
)

df_raw.loc[
    mask_5li,
    "gelir_5li_std"
] = gelir_5li_orj[mask_5li]

df_raw.loc[
    mask_5li,
    "gelir_kaynak"
] = "orijinal_5li"


print("\nGELİR KAYNAKLARI")
print(
    df_raw["gelir_kaynak"].value_counts(dropna=False)
)

print("\nGELİR 5'Lİ DAĞILIM")
print(
    df_raw["gelir_5li_std"]
    .value_counts(dropna=False)
    .sort_index()
)


# ============================================================
# 6) GELİR DUMMYLERİ
# REFERANS: GELİR 1
# ============================================================

df_raw["gelir2"] = np.where(
    df_raw["gelir_5li_std"] == 2,
    1,
    np.where(df_raw["gelir_5li_std"].notna(), 0, np.nan)
)

df_raw["gelir3"] = np.where(
    df_raw["gelir_5li_std"] == 3,
    1,
    np.where(df_raw["gelir_5li_std"].notna(), 0, np.nan)
)

df_raw["gelir4"] = np.where(
    df_raw["gelir_5li_std"] == 4,
    1,
    np.where(df_raw["gelir_5li_std"].notna(), 0, np.nan)
)

df_raw["gelir5"] = np.where(
    df_raw["gelir_5li_std"] == 5,
    1,
    np.where(df_raw["gelir_5li_std"].notna(), 0, np.nan)
)


# ============================================================
# 7) DİĞER BAĞIMSIZ DEĞİŞKENLER
# ============================================================

# ------------------------------------------------------------
# Cinsiyet
# REFERANS: ERKEK
# ------------------------------------------------------------

df_raw["kadin"] = np.where(
    df_raw["cinsiyet"] == 2,
    1,
    np.where(df_raw["cinsiyet"] == 1, 0, np.nan)
)

# ------------------------------------------------------------
# Yaş grupları
# REFERANS: 15-30
# ------------------------------------------------------------

df_raw["yas31_45"] = np.where(
    (df_raw["yas"] >= 31) &
    (df_raw["yas"] <= 45),
    1,
    np.where(df_raw["yas"].notna(), 0, np.nan)
)

df_raw["yas46_60"] = np.where(
    (df_raw["yas"] >= 46) &
    (df_raw["yas"] <= 60),
    1,
    np.where(df_raw["yas"].notna(), 0, np.nan)
)

df_raw["yas60ustu"] = np.where(
    df_raw["yas"] > 60,
    1,
    np.where(df_raw["yas"].notna(), 0, np.nan)
)


# ============================================================
# 8) YIL DUMMYLERİ
# REFERANS: 2015
# ============================================================

for year in [
    2016, 2017, 2018, 2019, 2020,
    2021, 2022, 2023, 2024, 2025
]:

    df_raw[f"yil_{year}"] = np.where(
        df_raw["yil"] == year,
        1,
        np.where(df_raw["yil"].notna(), 0, np.nan)
    )


# ============================================================
# 9) MODEL DEĞİŞKENLERİ
# ============================================================

X_COLS = [

    "kadin",

    "gelir2",
    "gelir3",
    "gelir4",
    "gelir5",

    "yas31_45",
    "yas46_60",
    "yas60ustu",

    "egitim_orta_lise",
    "egitim_universite",
    "egitim_uni_ustu",

    "yil_2016",
    "yil_2017",
    "yil_2018",
    "yil_2019",
    "yil_2020",
    "yil_2021",
    "yil_2022",
    "yil_2023",
    "yil_2024",
    "yil_2025"
]


# ============================================================
# 10) YARDIMCI FONKSİYONLAR
# ============================================================

def clean_X(X):

    X = X.copy()

    X = X.replace(
        [np.inf, -np.inf],
        np.nan
    )

    X = X.dropna(axis=1, how="all")

    nunique = X.nunique(dropna=True)

    drop_cols = nunique[
        nunique <= 1
    ].index.tolist()

    if len(drop_cols) > 0:

        print("\nModelden çıkarılan sabit değişkenler:")
        print(drop_cols)

    X = X.drop(
        columns=drop_cols,
        errors="ignore"
    )

    return X


def safe_summary_table(model):

    try:

        return pd.DataFrame(
            model.summary2().tables[1]
        )

    except Exception:

        conf = model.conf_int()

        return pd.DataFrame({

            "Coef.": model.params,
            "Std.Err.": model.bse,
            "z": model.tvalues,
            "P>|z|": model.pvalues,
            "[0.025": conf[0],
            "0.975]": conf[1]

        })


def safe_marginal_effects(model, X):

    try:

        margeff = model.get_margeff(
            at="overall"
        )

        arr = margeff.margeff
        se = margeff.margeff_se
        z = margeff.tvalues
        p = margeff.pvalues
        ci = margeff.conf_int()

        names = [
            c for c in X.columns
            if c != "const"
        ]

        min_len = min(
            len(names),
            len(arr),
            len(se),
            len(z),
            len(p),
            len(ci)
        )

        return pd.DataFrame({

            "Degisken": names[:min_len],
            "dy_dx": arr[:min_len],
            "Std_Hata": se[:min_len],
            "z": z[:min_len],
            "P_degeri": p[:min_len],
            "CI_alt": ci[:min_len, 0],
            "CI_ust": ci[:min_len, 1]

        })

    except Exception as e:

        return pd.DataFrame({
            "Hata": [str(e)]
        })


def run_model(dep_var, title):

    print("\n" + "=" * 70)
    print(title)
    print("=" * 70)

    X = df_raw[X_COLS].copy()

    y = df_raw[dep_var].copy()

    model_df = pd.concat(
        [X, y],
        axis=1
    )

    model_df = model_df.replace(
        [np.inf, -np.inf],
        np.nan
    )

    model_df = model_df.dropna()

    y = model_df[dep_var]

    X = model_df.drop(
        columns=[dep_var]
    )

    X = clean_X(X)

    X = sm.add_constant(
        X,
        has_constant="add"
    )

    print("Gözlem Sayısı:", len(model_df))

    print("\nBağımlı değişken dağılımı:")
    print(y.value_counts())

    try:

        model = Probit(y, X).fit(
            method="newton",
            maxiter=500,
            disp=False
        )

    except Exception as e:

        print("\nNewton hata verdi:")
        print(e)

        print("\nBFGS deneniyor...")

        model = Probit(y, X).fit(
            method="bfgs",
            maxiter=1000,
            disp=False
        )

    print(model.summary())

    summary_df = safe_summary_table(model)

    marginal_df = safe_marginal_effects(
        model,
        X
    )

    return (
        model,
        summary_df,
        marginal_df,
        X,
        y
    )


def calculate_vif(X):

    vif_X = X.copy()

    if "const" in vif_X.columns:

        vif_X = vif_X.drop(
            columns=["const"]
        )

    vif_X = vif_X.astype(float)

    vif_df = pd.DataFrame()

    vif_df["Degisken"] = vif_X.columns

    vif_df["VIF"] = [

        variance_inflation_factor(
            vif_X.values,
            i
        )

        for i in range(vif_X.shape[1])

    ]

    return vif_df


# ============================================================
# 11) MODELLER
# ============================================================

(
    model_online,
    online_summary,
    online_marginal,
    X_online,
    y_online
) = run_model(

    "saglik_online",
    "MODEL 1 - ONLINE SAĞLIK HİZMETİ"

)

(
    model_eticaret,
    eticaret_summary,
    eticaret_marginal,
    X_eticaret,
    y_eticaret
) = run_model(

    "ilac_eticaret",
    "MODEL 2 - E-TİCARET İLAÇ ALIŞVERİŞİ"

)


# ============================================================
# 12) VIF
# ============================================================

vif_online = calculate_vif(X_online)

vif_eticaret = calculate_vif(X_eticaret)

print("\nVIF ONLINE")
print(vif_online)

print("\nVIF ETICARET")
print(vif_eticaret)


# ============================================================
# 13) GRAFİK
# ============================================================

fig, axes = plt.subplots(
    1,
    2,
    figsize=(14, 5),
    sharey=True
)

MODELS = [

    (
        model_online,
        y_online,
        "Online Sağlık",
        axes[0]
    ),

    (
        model_eticaret,
        y_eticaret,
        "E-Ticaret İlaç",
        axes[1]
    )

]

for model, y, title, ax in MODELS:

    probs = model.predict()

    plot_df = pd.DataFrame({

        "yil": df_raw.loc[y.index, "yil"],
        "prob": probs,
        "y": y

    })

    agg = plot_df.groupby("yil").agg(

        tahmin=("prob", "mean"),
        gozlem=("y", "mean")

    )

    x_pos = range(len(agg))

    ax.bar(
        x_pos,
        agg["tahmin"],
        alpha=0.7,
        label="Tahmin"
    )

    ax.plot(
        list(x_pos),
        agg["gozlem"],
        "o-",
        linewidth=2,
        label="Gözlenen"
    )

    ax.set_xticks(list(x_pos))

    ax.set_xticklabels(
        agg.index.astype(int),
        rotation=45
    )

    ax.set_title(title)

    ax.set_ylim(0, 1)

    ax.legend()

plt.tight_layout()

plt.savefig(
    GRAFIK_PATH,
    dpi=150,
    bbox_inches="tight"
)

print("\nGrafik kaydedildi:")
print(GRAFIK_PATH)


# ============================================================
# 14) EXCEL'E AKTAR
# ============================================================

with pd.ExcelWriter(
    EXCEL_PATH,
    engine="openpyxl"
) as writer:

    online_summary.to_excel(
        writer,
        sheet_name="Online_Probit"
    )

    online_marginal.to_excel(
        writer,
        sheet_name="Online_Marjinal",
        index=False
    )

    eticaret_summary.to_excel(
        writer,
        sheet_name="Eticaret_Probit"
    )

    eticaret_marginal.to_excel(
        writer,
        sheet_name="Eticaret_Marjinal",
        index=False
    )

    vif_online.to_excel(
        writer,
        sheet_name="VIF_Online",
        index=False
    )

    vif_eticaret.to_excel(
        writer,
        sheet_name="VIF_Eticaret",
        index=False
    )

    df_raw[
        "gelir_kaynak"
    ].value_counts(dropna=False).reset_index().to_excel(

        writer,
        sheet_name="Gelir_Kaynagi",
        index=False

    )

    df_raw[
        "gelir_5li_std"
    ].value_counts(dropna=False).sort_index().reset_index().to_excel(

        writer,
        sheet_name="Gelir_Dagilim",
        index=False

    )

    df_raw[
        "yil"
    ].value_counts(dropna=False).sort_index().reset_index().to_excel(

        writer,
        sheet_name="Yillara_Gore_Gozlem",
        index=False

    )

    pd.DataFrame({

        "Degisken": X_online.columns,
        "Modelde_Kullanildi": "Evet"

    }).to_excel(

        writer,
        sheet_name="Model_Degiskenleri",
        index=False

    )


print("\n" + "=" * 70)
print("ANALİZ TAMAMLANDI")
print("=" * 70)

print("\nExcel:")
print(EXCEL_PATH)

print("\nGrafik:")
print(GRAFIK_PATH)
```
