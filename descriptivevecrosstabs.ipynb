```python
import pandas as pd
import numpy as np

# 1) Kolon isimlerini aynen kullanacağız
col_raw = "hane gelir aylık"
col_20  = "hanenin aylık toplam net geliri (20'li gelir grubu)"
col_5   = "hanenin aylık toplam net geliri (5'li gelir grubu)"

# 2) Yeni standart değişken
df["gelir_5li"] = np.nan

# 3) Önce hazır 5'li gelir grubunu al
df.loc[df[col_5].notna(), "gelir_5li"] = df.loc[df[col_5].notna(), col_5]

# 4) 5'li boşsa, 20'li gelir grubunu 5'liye çevir
mask_20 = df["gelir_5li"].isna() & df[col_20].notna()

df.loc[mask_20, "gelir_5li"] = (
    ((df.loc[mask_20, col_20].astype(float) - 1) // 4) + 1
)

# 5) Hâlâ boşsa, sayısal hane gelir aylık üzerinden 5'li grup oluştur
mask_raw = df["gelir_5li"].isna() & df[col_raw].notna()

df.loc[mask_raw, "gelir_5li"] = pd.qcut(
    df.loc[mask_raw, col_raw],
    q=5,
    labels=[1, 2, 3, 4, 5],
    duplicates="drop"
)

# 6) Sayısal formata çevir
df["gelir_5li"] = pd.to_numeric(df["gelir_5li"], errors="coerce")

# 7) Kontrol
print("Eksik gelir_5li sayısı:", df["gelir_5li"].isna().sum())
print(df["gelir_5li"].value_counts(dropna=False).sort_index())
```

```python
# df_clean'i yeniden oluştur
df_clean = df[df["gelir_5li"].notna()].copy()

# label kolonlarını df_clean içinde oluştur
df_clean["saglik_internet_label"] = df_clean["online sağlık hizmeti kullanımı"].map({
    1: "Kullanıyor",
    2: "Kullanmıyor"
})

df_clean["cinsiyet_label"] = df_clean["cinsiyet"].map({
    1: "Erkek",
    2: "Kadın"
})

# boşları çıkar
df_ct = df_clean[
    df_clean["saglik_internet_label"].notna() &
    df_clean["cinsiyet_label"].notna()
].copy()

# crosstab
ct = pd.crosstab(
    df_ct["gelir_5li"],
    [df_ct["saglik_internet_label"], df_ct["cinsiyet_label"]],
    normalize="index"
) * 100

ct.columns = [f"{a} {b}" for a, b in ct.columns]

# kolon eksikse hata vermesin
istenen_sira = [
    "Kullanıyor Erkek",
    "Kullanıyor Kadın",
    "Kullanmıyor Erkek",
    "Kullanmıyor Kadın"
]

ct = ct.reindex(columns=istenen_sira).fillna(0).round(2)

print(ct.to_string())
```
