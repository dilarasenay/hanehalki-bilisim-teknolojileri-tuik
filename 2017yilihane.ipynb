```python
import pandas as pd
df = pd.read_csv(r"C:\Users\dilara\Desktop\Hane halkı bilişim tüik\ferthane\yıl2017.csv")
df.head()
```

```python
# =========================================================
# ONLINE SAĞLIK HİZMETİ KULLANIMI (VAR / YOK)
# CİNSİYETE GÖRE FREKANS + YÜZDE TABLOSU
# =========================================================

import pandas as pd

# VAR
online_var = df[df["online sağlık hizmeti kullanımı"] == 1]["cinsiyet"] \
    .value_counts()

online_var_yuzde = round(
    df[df["online sağlık hizmeti kullanımı"] == 1]["cinsiyet"]
    .value_counts(normalize=True) * 100,
    2
)

# YOK
online_yok = df[df["online sağlık hizmeti kullanımı"] == 2]["cinsiyet"] \
    .value_counts()

online_yok_yuzde = round(
    df[df["online sağlık hizmeti kullanımı"] == 2]["cinsiyet"]
    .value_counts(normalize=True) * 100,
    2
)

# TABLO
online_tablo = pd.DataFrame({

    "online sağlık hizmeti kullanımı var": [
        online_var.get(2, 0),
        online_var.get(1, 0),
        online_var.sum()
    ],

    "online sağlık hizmeti kullanımı var yüzde": [
        online_var_yuzde.get(2, 0),
        online_var_yuzde.get(1, 0),
        100
    ],

    "online sağlık hizmeti kullanımı yok": [
        online_yok.get(2, 0),
        online_yok.get(1, 0),
        online_yok.sum()
    ],

    "online sağlık hizmeti kullanımı yok yüzde": [
        online_yok_yuzde.get(2, 0),
        online_yok_yuzde.get(1, 0),
        100
    ]

},

index=["kadın", "erkek", "toplam"]

)

display(online_tablo)
```

```python
# =========================================================
# ETİCARET İLAÇ ALIŞVERİŞİ (VAR / YOK)
# CİNSİYETE GÖRE FREKANS + YÜZDE TABLOSU
# =========================================================

import pandas as pd

# VAR
eticaret_var = df[df["eticaret ilaç alışverişi"] == 1]["cinsiyet"] \
    .value_counts()

eticaret_var_yuzde = round(
    df[df["eticaret ilaç alışverişi"] == 1]["cinsiyet"]
    .value_counts(normalize=True) * 100,
    2
)

# YOK
eticaret_yok = df[df["eticaret ilaç alışverişi"] == 2]["cinsiyet"] \
    .value_counts()

eticaret_yok_yuzde = round(
    df[df["eticaret ilaç alışverişi"] == 2]["cinsiyet"]
    .value_counts(normalize=True) * 100,
    2
)

# TABLO
eticaret_tablo = pd.DataFrame({

    "eticaret ilaç alışverişi var": [
        eticaret_var.get(2, 0),
        eticaret_var.get(1, 0),
        eticaret_var.sum()
    ],

    "eticaret ilaç alışverişi var yüzde": [
        eticaret_var_yuzde.get(2, 0),
        eticaret_var_yuzde.get(1, 0),
        100
    ],

    "eticaret ilaç alışverişi yok": [
        eticaret_yok.get(2, 0),
        eticaret_yok.get(1, 0),
        eticaret_yok.sum()
    ],

    "eticaret ilaç alışverişi yok yüzde": [
        eticaret_yok_yuzde.get(2, 0),
        eticaret_yok_yuzde.get(1, 0),
        100
    ]

},

index=["kadın", "erkek", "toplam"]

)

display(eticaret_tablo)
```

```python
import pandas as pd

# ======================================================
# GELİR GRUBU
# ======================================================

df = df.sort_values("hane gelir aylık").reset_index(drop=True)

df["gelir_grup"] = pd.qcut(
    df.index,
    q=5,
    labels=[1, 2, 3, 4, 5]
)

# ======================================================
# ONLINE SAĞLIK HİZMETİ KULLANIMI
# 1 = var, 2 = yok
# cinsiyet: 1 = erkek, 2 = kadın
# ======================================================

sonuc = []

for gelir in [1, 2, 3, 4, 5]:

    alt_df = df[df["gelir_grup"] == gelir]

    var_df = alt_df[alt_df["online sağlık hizmeti kullanımı"] == 1]
    yok_df = alt_df[alt_df["online sağlık hizmeti kullanımı"] == 2]

    var_cinsiyet_yuzde = var_df["cinsiyet"].value_counts(normalize=True) * 100
    yok_cinsiyet_yuzde = yok_df["cinsiyet"].value_counts(normalize=True) * 100

    sonuc.append({
        "yıl": 2017,
        "gelir": gelir,

        "var kadın": round(var_cinsiyet_yuzde.get(2, 0), 2),
        "var erkek": round(var_cinsiyet_yuzde.get(1, 0), 2),
        "var toplam": len(var_df),

        "yok kadın": round(yok_cinsiyet_yuzde.get(2, 0), 2),
        "yok erkek": round(yok_cinsiyet_yuzde.get(1, 0), 2),
        "yok toplam": len(yok_df)
    })

gelir_online_saglik_2017 = pd.DataFrame(sonuc)

# ======================================================
# EN ALT FREKANS SATIRI
# kadın/erkek/toplam hepsi frekans
# ======================================================

toplam_var = df[df["online sağlık hizmeti kullanımı"] == 1]
toplam_yok = df[df["online sağlık hizmeti kullanımı"] == 2]

toplam_var_cinsiyet = toplam_var["cinsiyet"].value_counts()
toplam_yok_cinsiyet = toplam_yok["cinsiyet"].value_counts()

toplam_satiri = pd.DataFrame([{
    "yıl": 2017,
    "gelir": "frekans",

    "var kadın": toplam_var_cinsiyet.get(2, 0),
    "var erkek": toplam_var_cinsiyet.get(1, 0),
    "var toplam": len(toplam_var),

    "yok kadın": toplam_yok_cinsiyet.get(2, 0),
    "yok erkek": toplam_yok_cinsiyet.get(1, 0),
    "yok toplam": len(toplam_yok)
}])

gelir_online_saglik_2017= pd.concat(
    [gelir_online_saglik_2017, toplam_satiri],
    ignore_index=True
)

display(gelir_online_saglik_2017)
```

```python
# =========================================================
# 2015 - ETİCARETTEN İLAÇ ALIŞVERİŞİ
# GELİR GRUBU + CİNSİYET TABLOSU
# =========================================================

import pandas as pd

# =========================================================
# GELİR GRUBU OLUŞTUR
# =========================================================

df = df.sort_values(
    "hane gelir aylık"
).reset_index(drop=True)

df["gelir_grup"] = pd.qcut(
    df.index,
    q=5,
    labels=[1, 2, 3, 4, 5]
)

# =========================================================
# SONUÇ
# =========================================================

sonuc = []

for grup in [1, 2, 3, 4, 5]:

    grup_df = df[df["gelir_grup"] == grup]

    # VAR
    var_df = grup_df[
        grup_df["eticaret ilaç alışverişi"] == 1
    ]

    var_cinsiyet = (
        var_df["cinsiyet"]
        .value_counts(normalize=True) * 100
    ).round(2)

    # YOK
    yok_df = grup_df[
        grup_df["eticaret ilaç alışverişi"] == 2
    ]

    yok_cinsiyet = (
        yok_df["cinsiyet"]
        .value_counts(normalize=True) * 100
    ).round(2)

    sonuc.append({

        "yıl": 2015,

        "gelir": grup,

        # VAR
        "var kadın":
            var_cinsiyet.get(2, 0),

        "var erkek":
            var_cinsiyet.get(1, 0),

        "var toplam":
            len(var_df),

        # YOK
        "yok kadın":
            yok_cinsiyet.get(2, 0),

        "yok erkek":
            yok_cinsiyet.get(1, 0),

        "yok toplam":
            len(yok_df)
    })

# =========================================================
# TOPLAM SATIRI
# =========================================================

toplam_satiri = pd.DataFrame([{

    "yıl": 2015,

    "gelir": "toplam",

    "var kadın":
        len(df[
            (df["eticaret ilaç alışverişi"] == 1) &
            (df["cinsiyet"] == 2)
        ]),

    "var erkek":
        len(df[
            (df["eticaret ilaç alışverişi"] == 1) &
            (df["cinsiyet"] == 1)
        ]),

    "var toplam":
        len(df[df["eticaret ilaç alışverişi"] == 1]),

    "yok kadın":
        len(df[
            (df["eticaret ilaç alışverişi"] == 2) &
            (df["cinsiyet"] == 2)
        ]),

    "yok erkek":
        len(df[
            (df["eticaret ilaç alışverişi"] == 2) &
            (df["cinsiyet"] == 1)
        ]),

    "yok toplam":
        len(df[df["eticaret ilaç alışverişi"] == 2])

}])

# =========================================================
# BİRLEŞTİR
# =========================================================

gelir_eticaret_2015 = pd.concat(
    [pd.DataFrame(sonuc), toplam_satiri],
    ignore_index=True
)

display(gelir_eticaret_2015)
```

```python
# =========================================================
# YAŞ GRUBU x ONLINE SAĞLIK HİZMETİ KULLANIMI
# CİNSİYETE GÖRE FREKANS + YÜZDE TABLOSU
# =========================================================

import pandas as pd

# =========================================================
# YAŞ GRUBU OLUŞTUR
# =========================================================

df["yas_grup"] = pd.cut(
    df["yaş"],
    bins=[15, 30, 45, 60, 100],
    labels=["15-30", "31-45", "46-60", "60+"],
    include_lowest=True
)

# =========================================================
# TABLO
# =========================================================

sonuc = []

for yas in ["15-30", "31-45", "46-60", "60+"]:

    alt = df[df["yas_grup"] == yas]

    # ONLINE SAĞLIK HİZMETİ KULLANIMI VAR
    online_var = alt[
        alt["online sağlık hizmeti kullanımı"] == 1
    ]

    online_var_cinsiyet = (
        online_var["cinsiyet"]
        .value_counts(normalize=True) * 100
    )

    # ONLINE SAĞLIK HİZMETİ KULLANIMI YOK
    online_yok = alt[
        alt["online sağlık hizmeti kullanımı"] == 2
    ]

    online_yok_cinsiyet = (
        online_yok["cinsiyet"]
        .value_counts(normalize=True) * 100
    )

    sonuc.append({

        "yıl": 2015,

        "yaş": yas,

        # VAR
        "var kadın":
            round(online_var_cinsiyet.get(2, 0), 2),

        "var erkek":
            round(online_var_cinsiyet.get(1, 0), 2),

        "var toplam":
            len(online_var),

        # YOK
        "yok kadın":
            round(online_yok_cinsiyet.get(2, 0), 2),

        "yok erkek":
            round(online_yok_cinsiyet.get(1, 0), 2),

        "yok toplam":
            len(online_yok)

    })

# =========================================================
# FREKANS SATIRI
# =========================================================

toplam_satiri = pd.DataFrame([{

    "yıl": 2015,

    "yaş": "toplam",

    "var kadın":
        df[df["online sağlık hizmeti kullanımı"] == 1]["cinsiyet"]
        .value_counts()
        .get(2, 0),

    "var erkek":
        df[df["online sağlık hizmeti kullanımı"] == 1]["cinsiyet"]
        .value_counts()
        .get(1, 0),

    "var toplam":
        len(df[df["online sağlık hizmeti kullanımı"] == 1]),

    "yok kadın":
        df[df["online sağlık hizmeti kullanımı"] == 2]["cinsiyet"]
        .value_counts()
        .get(2, 0),

    "yok erkek":
        df[df["online sağlık hizmeti kullanımı"] == 2]["cinsiyet"]
        .value_counts()
        .get(1, 0),

    "yok toplam":
        len(df[df["online sağlık hizmeti kullanımı"] == 2])

}])

# =========================================================
# BİRLEŞTİR
# =========================================================

yas_online_2015 = pd.concat(
    [pd.DataFrame(sonuc), toplam_satiri],
    ignore_index=True
)

display(yas_online_2015)
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm

# =====================================================
# MODEL 2 - PROBIT MODEL
# Bağımlı değişken: E-ticaretten ilaç alışverişi
# Gelir değişkeni: hane gelir aylık
# =====================================================

# 1) BAĞIMLI DEĞİŞKEN
df["eticaret_ilac"] = df["eticaret ilaç alışverişi"].replace({
    "Evet": 1, "Hayır": 0,
    "evet": 1, "hayır": 0,
    "EVET": 1, "HAYIR": 0,
    1: 1,
    2: 0,
    0: 0
})

# 2) CİNSİYET
df["cinsiyet_kadin"] = df["cinsiyet"].replace({
    "Erkek": 0, "Kadın": 1,
    "erkek": 0, "kadın": 1,
    "ERKEK": 0, "KADIN": 1,
    1: 0,
    2: 1
})

# 3) YAŞ KATEGORİLERİ
df["yas"] = pd.to_numeric(df["yaş"], errors="coerce")

df["yas_grup"] = pd.cut(
    df["yas"],
    bins=[14, 30, 45, 60, 120],
    labels=["15-30", "31-45", "46-60", "60+"]
)

yas_dummy = pd.get_dummies(df["yas_grup"], prefix="yas")
yas_dummy = yas_dummy.drop(columns=["yas_15-30"], errors="ignore")

# 4) EĞİTİM KATEGORİLERİ
df["egitim_sayi"] = pd.to_numeric(df["eğitim"], errors="coerce")

def egitim_kategori(x):
    if x in [0, 1]:
        return "Ilkokul_ve_alti"
    elif x in [2, 3]:
        return "Orta_Lise"
    elif x in [4, 5, 6]:
        return "Universite"
    elif x == 7:
        return "Doktora"
    else:
        return np.nan

df["egitim_grup"] = df["egitim_sayi"].apply(egitim_kategori)

egitim_dummy = pd.get_dummies(df["egitim_grup"], prefix="egitim")
egitim_dummy = egitim_dummy.drop(
    columns=["egitim_Ilkokul_ve_alti"],
    errors="ignore"
)

# 5) GELİR
df["gelir"] = pd.to_numeric(
    df["hane gelir aylık"],
    errors="coerce"
)

df["log_gelir"] = np.log1p(df["gelir"])

# 6) MODEL DATAFRAME
model_df = pd.concat(
    [
        df[["eticaret_ilac", "cinsiyet_kadin", "log_gelir"]],
        yas_dummy,
        egitim_dummy
    ],
    axis=1
)

# 7) SAYISALA ÇEVİR VE TEMİZLE
model_df = model_df.apply(pd.to_numeric, errors="coerce")

print("Dropna öncesi satır sayısı:", len(model_df))
print("\nEksik değer sayıları:")
print(model_df.isna().sum())

model_df = model_df.dropna()

print("\nDropna sonrası satır sayısı:", len(model_df))

# 8) X ve y
y = model_df["eticaret_ilac"]
X = model_df.drop(columns=["eticaret_ilac"])

X = sm.add_constant(X)

X = X.astype(float)
y = y.astype(float)

# 9) KONTROLLER
print("\nBağımlı değişken dağılımı:")
print(y.value_counts())

print("\nEğitim grubu dağılımı:")
print(df["egitim_grup"].value_counts(dropna=False))

print("\nGelir dağılımı:")
print(df["gelir"].value_counts(dropna=False).head(20))

print("\nModel değişkenleri:")
print(X.columns)

# 10) PROBIT MODEL
probit_model = sm.Probit(y, X)
result = probit_model.fit()

print(result.summary().as_text())
```

```python
# =====================================================
# ETİCARET İLAÇ ALIŞVERİŞİ - PROBIT MODEL
# 2018 UYUMLU FINAL KOD
# =====================================================

import pandas as pd
import numpy as np
import statsmodels.api as sm

# -----------------------------------------------------
# BAĞIMLI DEĞİŞKEN
# ETİCARET İLAÇ ALIŞVERİŞİ
# -----------------------------------------------------

df["eticaret_ilac"] = df["eticaret ilaç alışverişi"].replace({
    "Evet": 1,
    "Hayır": 0,
    "evet": 1,
    "hayır": 0,
    1: 1,
    2: 0,
    0: 0
})

# -----------------------------------------------------
# CİNSİYET
# Erkek = 0
# Kadın = 1
# -----------------------------------------------------

df["cinsiyet_kadin"] = df["cinsiyet"].replace({
    "Erkek": 0,
    "Kadın": 1,
    "erkek": 0,
    "kadın": 1,
    1: 0,
    2: 1
})

# -----------------------------------------------------
# YAŞ
# -----------------------------------------------------

df["yas"] = pd.to_numeric(
    df["yaş"],
    errors="coerce"
)

df["yas_grup"] = pd.cut(
    df["yas"],
    bins=[14, 30, 45, 60, 120],
    labels=[
        "15-30",
        "31-45",
        "46-60",
        "60+"
    ]
)

yas_dummy = pd.get_dummies(
    df["yas_grup"],
    prefix="yas"
)

# REFERANS GRUP
yas_dummy = yas_dummy.drop(
    columns=["yas_15-30"],
    errors="ignore"
)

# -----------------------------------------------------
# EĞİTİM
# 2018 UYUMLU
# -----------------------------------------------------

df["egitim_sayi"] = pd.to_numeric(
    df["eğitim"],
    errors="coerce"
)

def egitim_kategori(x):

    # REFERANS
    if x in [1, 2]:
        return "Ilkokul_ve_alti"

    # ORTAOKUL / LİSE
    elif x in [3, 4]:
        return "Ortaokul_Lise"

    # ÜNİVERSİTE
    elif x in [51, 511, 512]:
        return "Universite"

    # ÜNİVERSİTEÜSTÜ
    elif x in [52, 53]:
        return "Universiteustu"

    else:
        return np.nan

df["egitim_grup"] = df["egitim_sayi"].apply(
    egitim_kategori
)

egitim_dummy = pd.get_dummies(
    df["egitim_grup"],
    prefix="egitim"
)

# REFERANS:
# İlkokul ve altı

egitim_dummy = egitim_dummy.drop(
    columns=["egitim_Ilkokul_ve_alti"],
    errors="ignore"
)

# -----------------------------------------------------
# GELİR
# -----------------------------------------------------

df["gelir"] = pd.to_numeric(
    df["hane gelir aylık"],
    errors="coerce"
)

# LOG GELİR
df["log_gelir"] = np.log1p(df["gelir"])

# -----------------------------------------------------
# MODEL DATAFRAME
# -----------------------------------------------------

model_df = pd.concat(
    [
        df[
            [
                "eticaret_ilac",
                "cinsiyet_kadin",
                "log_gelir"
            ]
        ],
        yas_dummy,
        egitim_dummy
    ],
    axis=1
)

# -----------------------------------------------------
# TÜM DEĞİŞKENLERİ SAYISALA ÇEVİR
# -----------------------------------------------------

model_df = model_df.apply(
    pd.to_numeric,
    errors="coerce"
)

# -----------------------------------------------------
# EKSİK GÖZLEMLERİ SİL
# -----------------------------------------------------

model_df = model_df.dropna()

# -----------------------------------------------------
# X ve y
# -----------------------------------------------------

y = model_df["eticaret_ilac"]

X = model_df.drop(
    columns=["eticaret_ilac"]
)

# CONSTANT
X = sm.add_constant(X)

# FLOAT
X = X.astype(float)
y = y.astype(float)

# -----------------------------------------------------
# MODEL
# -----------------------------------------------------

probit_model = sm.Probit(y, X)

result = probit_model.fit()

# -----------------------------------------------------
# SONUÇLAR
# -----------------------------------------------------

summary_text = result.summary().as_text()

print(summary_text)

# -----------------------------------------------------
# MARJİNAL ETKİLER
# -----------------------------------------------------

```
