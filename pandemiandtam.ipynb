### 2015 yılı korelasyon 

```python
yıl2015 = pd.read_csv(r"C:\Users\dilara\Desktop\Hane halkı bilişim tüik\ferthane\yıl2015.csv")
yıl2015.head()
```

```python
import pandas as pd

#evet=1, hayır=0 
yıl2015["online_saglik"] = yıl2015["online sağlık hizmeti kullanımı"].replace({
    "Evet": 1,
    "Hayır": 0,
    1: 1,
    2: 0
})

yıl2015["ilac_eticaret"] = yıl2015["eticaret ilaç alışverişi"].replace({
    "Evet": 1,
    "Hayır": 0,
    1: 1,
    2: 0
})

# eksikler
corr_df = yıl2015[["online_saglik", "ilac_eticaret"]].dropna()

# korelasyon
corr = corr_df.corr()

print(corr)
```

```python
corr_value = corr_df["online_saglik"].corr(
    corr_df["ilac_eticaret"]
)

print("Korelasyon:", corr_value)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

yıl2015["online_saglik"] = yıl2015["online sağlık hizmeti kullanımı"].replace({
    "Evet": 1,
    "Hayır": 0,
    1: 1,
    2: 0
})

yıl2015["ilac_eticaret"] = yıl2015["eticaret ilaç alışverişi"].replace({
    "Evet": 1,
    "Hayır": 0,
    1: 1,
    2: 0
})

corr_df = yıl2015[
    ["online_saglik", "ilac_eticaret"]
].dropna()

corr_matrix = corr_df.corr()

sns.set_theme(style="white")

fig, ax = plt.subplots(figsize=(7,6))

# Modern soft palette
cmap = sns.color_palette(
    ["#EAFBF6", "#B8F2E6", "#5DD39E", "#348AA7"],
    as_cmap=True
)

sns.heatmap(
    corr_matrix,
    annot=True,
    fmt=".2f",
    cmap=cmap,
    vmin=-1,
    vmax=1,
    square=True,
    linewidths=4,
    linecolor="white",
    cbar=True,
    annot_kws={
        "fontsize":16,
        "fontweight":"bold",
        "color":"white"
    },
    ax=ax
)

ax.set_xticklabels(
    ["Online Sağlık", "E-Ticaret İlaç"],
    fontsize=12,
    fontweight="bold"
)

ax.set_yticklabels(
    ["Online Sağlık", "E-Ticaret İlaç"],
    fontsize=12,
    fontweight="bold",
    rotation=0
)

plt.title(
    "Online Sağlık Hizmeti Kullanımı ve\n"
    "E-Ticaretten İlaç Alışverişi İlişkisi",
    fontsize=18,
    fontweight="bold",
    pad=24
)

for spine in ax.spines.values():
    spine.set_visible(False)

plt.tight_layout()
plt.show()
```
