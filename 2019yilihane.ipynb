```python
# =========================================================
# ONLINE SAĞLIK HİZMETİ KULLANIMI (VAR / YOK)
# CİNSİYETE GÖRE FREKANS + YÜZDE TABLOSU
# =========================================================

import pandas as pd

# VAR
online_var = df[df["sağlık bilgisi arama"] == 1]["cinsiyet"] \
    .value_counts()

online_var_yuzde = round(
    df[df["sağlık bilgisi arama"] == 1]["cinsiyet"]
    .value_counts(normalize=True) * 100,
    2
)

# YOK
online_yok = df[df["sağlık bilgisi arama"] == 2]["cinsiyet"] \
    .value_counts()

online_yok_yuzde = round(
    df[df["sağlık bilgisi arama"] == 2]["cinsiyet"]
    .value_counts(normalize=True) * 100,
    2
)

# TABLO
online_tablo = pd.DataFrame({

    "online sağlık hizmeti kullanımı var": [
        online_var.get(2, 0),
        online_var.get(1, 0),
        online_var.sum()
    ],

    "online sağlık hizmeti kullanımı var yüzde": [
        online_var_yuzde.get(2, 0),
        online_var_yuzde.get(1, 0),
        100
    ],

    "online sağlık hizmeti kullanımı yok": [
        online_yok.get(2, 0),
        online_yok.get(1, 0),
        online_yok.sum()
    ],

    "online sağlık hizmeti kullanımı yok yüzde": [
        online_yok_yuzde.get(2, 0),
        online_yok_yuzde.get(1, 0),
        100
    ]

},

index=["kadın", "erkek", "toplam"]

)

display(online_tablo)
```

```python
import pandas as pd

# ======================================================
# GELİR GRUBU
# ======================================================

df = df.sort_values("hane gelir aylık").reset_index(drop=True)

df["gelir_grup"] = pd.qcut(
    df.index,
    q=5,
    labels=[1, 2, 3, 4, 5]
)

# ======================================================
# ONLINE SAĞLIK HİZMETİ KULLANIMI
# 1 = var, 2 = yok
# cinsiyet: 1 = erkek, 2 = kadın
# ======================================================

sonuc = []

for gelir in [1, 2, 3, 4, 5]:

    alt_df = df[df["gelir_grup"] == gelir]

    var_df = alt_df[alt_df["sağlık bilgisi arama"] == 1]
    yok_df = alt_df[alt_df["sağlık bilgisi arama"] == 2]

    var_cinsiyet_yuzde = var_df["cinsiyet"].value_counts(normalize=True) * 100
    yok_cinsiyet_yuzde = yok_df["cinsiyet"].value_counts(normalize=True) * 100

    sonuc.append({
        "yıl": 2019,
        "gelir": gelir,

        "var kadın": round(var_cinsiyet_yuzde.get(2, 0), 2),
        "var erkek": round(var_cinsiyet_yuzde.get(1, 0), 2),
        "var toplam": len(var_df),

        "yok kadın": round(yok_cinsiyet_yuzde.get(2, 0), 2),
        "yok erkek": round(yok_cinsiyet_yuzde.get(1, 0), 2),
        "yok toplam": len(yok_df)
    })

gelir_online_saglik_2019 = pd.DataFrame(sonuc)

# ======================================================
# EN ALT FREKANS SATIRI
# kadın/erkek/toplam hepsi frekans
# ======================================================

toplam_var = df[df["sağlık bilgisi arama"] == 1]
toplam_yok = df[df["sağlık bilgisi arama"] == 2]

toplam_var_cinsiyet = toplam_var["cinsiyet"].value_counts()
toplam_yok_cinsiyet = toplam_yok["cinsiyet"].value_counts()

toplam_satiri = pd.DataFrame([{
    "yıl": 2019,
    "gelir": "frekans",

    "var kadın": toplam_var_cinsiyet.get(2, 0),
    "var erkek": toplam_var_cinsiyet.get(1, 0),
    "var toplam": len(toplam_var),

    "yok kadın": toplam_yok_cinsiyet.get(2, 0),
    "yok erkek": toplam_yok_cinsiyet.get(1, 0),
    "yok toplam": len(toplam_yok)
}])

gelir_online_saglik_2019= pd.concat(
    [gelir_online_saglik_2019, toplam_satiri],
    ignore_index=True
)

display(gelir_online_saglik_2019)
```

```python
import pandas as pd
import numpy as np

# =========================================================
# EĞİTİM KATEGORİLERİ (2017)
# =========================================================

def egitim_kategori(x):

    if pd.isna(x):
        return None

    x = int(x)

    # İLKOKUL VE ALTI
    if x in [1, 2]:
        return "ilkokulvealtı"

    # ORTAOKUL / LİSE
    elif x in [3, 4]:
        return "ortaokul/lise"

    # ÜNİVERSİTE
    elif x in [51, 52]:
        return "üniversite"

    # ÜNİVERSİTE ÜSTÜ
    elif x in [53]:
        return "üniversiteüstü"

    else:
        return None


# =========================================================
# EĞİTİM GRUBU OLUŞTUR
# =========================================================

df["egitim_grup"] = df["eğitim"].apply(egitim_kategori)

# =========================================================
# SONUÇ LİSTESİ
# =========================================================

sonuc = []

for grup in [
    "ilkokulvealtı",
    "ortaokul/lise",
    "üniversite",
    "üniversiteüstü"
]:

    temp = df[df["egitim_grup"] == grup]

    # =====================================================
    # SAĞLIK BİLGİSİ ARAMA = VAR
    # =====================================================

    var_df = temp[temp["sağlık bilgisi arama"] == 1]

    var_kadin = round(
        (var_df["cinsiyet"].value_counts(normalize=True).get(2, 0)) * 100,
        2
    )

    var_erkek = round(
        (var_df["cinsiyet"].value_counts(normalize=True).get(1, 0)) * 100,
        2
    )

    var_toplam = len(var_df)

    # =====================================================
    # SAĞLIK BİLGİSİ ARAMA = YOK
    # =====================================================

    yok_df = temp[temp["sağlık bilgisi arama"] == 2]

    yok_kadin = round(
        (yok_df["cinsiyet"].value_counts(normalize=True).get(2, 0)) * 100,
        2
    )

    yok_erkek = round(
        (yok_df["cinsiyet"].value_counts(normalize=True).get(1, 0)) * 100,
        2
    )

    yok_toplam = len(yok_df)

    sonuc.append({
        "yıl": 2017,
        "eğitim": grup,
        "var kadın": var_kadin,
        "var erkek": var_erkek,
        "var toplam": var_toplam,
        "yok kadın": yok_kadin,
        "yok erkek": yok_erkek,
        "yok toplam": yok_toplam
    })

# =========================================================
# TOPLAM SATIRI
# =========================================================

var_all = df[df["sağlık bilgisi arama"] == 1]
yok_all = df[df["sağlık bilgisi arama"] == 2]

toplam_satiri = pd.DataFrame([{
    "yıl": 2017,
    "eğitim": "toplam",

    "var kadın": len(var_all[var_all["cinsiyet"] == 2]),
    "var erkek": len(var_all[var_all["cinsiyet"] == 1]),
    "var toplam": len(var_all),

    "yok kadın": len(yok_all[yok_all["cinsiyet"] == 2]),
    "yok erkek": len(yok_all[yok_all["cinsiyet"] == 1]),
    "yok toplam": len(yok_all)
}])

# =========================================================
# BİRLEŞTİR
# =========================================================

egitim_saglik_2017 = pd.concat(
    [pd.DataFrame(sonuc), toplam_satiri],
    ignore_index=True
)

display(egitim_saglik_2017)
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm

# =====================================================
# MODEL 1 - PROBIT MODEL (2017)
# Bağımlı değişken:
# Online sağlık hizmeti kullanımı
#
# Referans Gruplar:
# Cinsiyet -> Erkek
# Yaş -> 15-30
# Eğitim -> İlkokul ve altı
# =====================================================

# -----------------------------------------------------
# 1) ONLINE SAĞLIK HİZMETİ KULLANIMI
# -----------------------------------------------------

df["online_saglik"] = df["sağlık bilgisi arama"].replace({
    "Evet": 1,
    "Hayır": 0,
    "evet": 1,
    "hayır": 0,
    1: 1,
    2: 0,
    0: 0
})

# -----------------------------------------------------
# 2) CİNSİYET
# Erkek = 0
# Kadın = 1
# -----------------------------------------------------

df["cinsiyet_kadin"] = df["cinsiyet"].replace({
    "Erkek": 0,
    "Kadın": 1,
    "erkek": 0,
    "kadın": 1,
    1: 0,
    2: 1
})

# -----------------------------------------------------
# 3) YAŞ KATEGORİLERİ
# -----------------------------------------------------

df["yas"] = pd.to_numeric(
    df["yaş"],
    errors="coerce"
)

df["yas_grup"] = pd.cut(
    df["yas"],
    bins=[14, 30, 45, 60, 120],
    labels=[
        "15-30",
        "31-45",
        "46-60",
        "60+"
    ]
)

yas_dummy = pd.get_dummies(
    df["yas_grup"],
    prefix="yas"
)

# REFERANS YAŞ GRUBU
yas_dummy = yas_dummy.drop(
    columns=["yas_15-30"],
    errors="ignore"
)

# -----------------------------------------------------
# 4) EĞİTİM KATEGORİLERİ (2017 yapısı)
# -----------------------------------------------------

df["egitim_sayi"] = pd.to_numeric(
    df["eğitim"],
    errors="coerce"
)

def egitim_kategori(x):

    # REFERANS GRUP
    if x in [1]:
        return "Ilkokul_ve_alti"

    elif x in [2, 3, 4]:
        return "Orta_Lise"

    elif x in [51, 52]:
        return "Universite"

    elif x == 53:
        return "Doktora"

    else:
        return np.nan

df["egitim_grup"] = df["egitim_sayi"].apply(
    egitim_kategori
)

egitim_dummy = pd.get_dummies(
    df["egitim_grup"],
    prefix="egitim"
)

# REFERANS EĞİTİM GRUBU
egitim_dummy = egitim_dummy.drop(
    columns=["egitim_Ilkokul_ve_alti"],
    errors="ignore"
)

# -----------------------------------------------------
# 5) GELİR
# -----------------------------------------------------

df["gelir"] = pd.to_numeric(
    df["hane gelir aylık"],
    errors="coerce"
)

# LOG GELİR
df["log_gelir"] = np.log1p(df["gelir"])

# -----------------------------------------------------
# 6) MODEL DATAFRAME
# -----------------------------------------------------

model_df = pd.concat(
    [
        df[
            [
                "online_saglik",
                "cinsiyet_kadin",
                "log_gelir"
            ]
        ],
        yas_dummy,
        egitim_dummy
    ],
    axis=1
)

# -----------------------------------------------------
# 7) TÜM DEĞİŞKENLERİ SAYISALA ÇEVİR
# -----------------------------------------------------

model_df = model_df.apply(
    pd.to_numeric,
    errors="coerce"
)

# Eksik gözlemleri sil
model_df = model_df.dropna()

# -----------------------------------------------------
# 8) X ve y
# -----------------------------------------------------

y = model_df["online_saglik"]

X = model_df.drop(
    columns=["online_saglik"]
)

# CONSTANT
X = sm.add_constant(X)

# FLOAT
X = X.astype(float)
y = y.astype(float)

# -----------------------------------------------------
# 9) KONTROLLER
# -----------------------------------------------------

print("Modelde kullanılan gözlem sayısı:")
print(len(model_df))

print("\nBağımlı değişken dağılımı:")
print(y.value_counts())

print("\nModel değişkenleri:")
print(X.columns)

# -----------------------------------------------------
# 10) PROBIT MODEL
# -----------------------------------------------------

probit_model = sm.Probit(y, X)

result = probit_model.fit()

# -----------------------------------------------------
# 11) SONUÇLAR
# -----------------------------------------------------

summary_text = result.summary().as_text()

print(summary_text)
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm

# =====================================================
# MODEL 1 - PROBIT MODEL (2017)
# Bağımlı değişken:
# Online sağlık hizmeti kullanımı
#
# Referans Gruplar:
# Cinsiyet -> Erkek
# Yaş -> 15-30
# Eğitim -> İlkokul ve altı
# =====================================================

# -----------------------------------------------------
# 1) ONLINE SAĞLIK HİZMETİ KULLANIMI
# -----------------------------------------------------

df["online_saglik"] = df["sağlık bilgisi arama"].replace({
    "Evet": 1,
    "Hayır": 0,
    "evet": 1,
    "hayır": 0,
    1: 1,
    2: 0,
    0: 0
})

# -----------------------------------------------------
# 2) CİNSİYET
# Erkek = 0
# Kadın = 1
# -----------------------------------------------------

df["cinsiyet_kadin"] = df["cinsiyet"].replace({
    "Erkek": 0,
    "Kadın": 1,
    "erkek": 0,
    "kadın": 1,
    1: 0,
    2: 1
})

# -----------------------------------------------------
# 3) YAŞ KATEGORİLERİ
# -----------------------------------------------------

df["yas"] = pd.to_numeric(
    df["yaş"],
    errors="coerce"
)

df["yas_grup"] = pd.cut(
    df["yas"],
    bins=[14, 30, 45, 60, 120],
    labels=[
        "15-30",
        "31-45",
        "46-60",
        "60+"
    ]
)

yas_dummy = pd.get_dummies(
    df["yas_grup"],
    prefix="yas"
)

# REFERANS YAŞ GRUBU
yas_dummy = yas_dummy.drop(
    columns=["yas_15-30"],
    errors="ignore"
)

# -----------------------------------------------------
# 4) EĞİTİM KATEGORİLERİ (2017 yapısı)
# -----------------------------------------------------

df["egitim_sayi"] = pd.to_numeric(
    df["eğitim"],
    errors="coerce"
)

def egitim_kategori(x):

    # REFERANS GRUP
    if x in [1]:
        return "Ilkokul_ve_alti"

    elif x in [2, 3, 4]:
        return "Orta_Lise"

    elif x in [51, 52]:
        return "Universite"

    elif x == 53:
        return "Doktora"

    else:
        return np.nan

df["egitim_grup"] = df["egitim_sayi"].apply(
    egitim_kategori
)

egitim_dummy = pd.get_dummies(
    df["egitim_grup"],
    prefix="egitim"
)

# REFERANS EĞİTİM GRUBU
egitim_dummy = egitim_dummy.drop(
    columns=["egitim_Ilkokul_ve_alti"],
    errors="ignore"
)

# -----------------------------------------------------
# 5) GELİR
# -----------------------------------------------------

df["gelir"] = pd.to_numeric(
    df["hane gelir aylık"],
    errors="coerce"
)

# LOG GELİR
df["log_gelir"] = np.log1p(df["gelir"])

# -----------------------------------------------------
# 6) MODEL DATAFRAME
# -----------------------------------------------------

model_df = pd.concat(
    [
        df[
            [
                "online_saglik",
                "cinsiyet_kadin",
                "log_gelir"
            ]
        ],
        yas_dummy,
        egitim_dummy
    ],
    axis=1
)

# -----------------------------------------------------
# 7) TÜM DEĞİŞKENLERİ SAYISALA ÇEVİR
# -----------------------------------------------------

model_df = model_df.apply(
    pd.to_numeric,
    errors="coerce"
)

# Eksik gözlemleri sil
model_df = model_df.dropna()

# -----------------------------------------------------
# 8) X ve y
# -----------------------------------------------------

y = model_df["online_saglik"]

X = model_df.drop(
    columns=["online_saglik"]
)

# CONSTANT
X = sm.add_constant(X)

# FLOAT
X = X.astype(float)
y = y.astype(float)

# -----------------------------------------------------
# 9) KONTROLLER
# -----------------------------------------------------

print("Modelde kullanılan gözlem sayısı:")
print(len(model_df))

print("\nBağımlı değişken dağılımı:")
print(y.value_counts())

print("\nModel değişkenleri:")
print(X.columns)

# -----------------------------------------------------
# 10) PROBIT MODEL
# -----------------------------------------------------

probit_model = sm.Probit(y, X)

result = probit_model.fit()

# -----------------------------------------------------
# 11) SONUÇLAR
# -----------------------------------------------------

summary_text = result.summary().as_text()

print(summary_text)
```
