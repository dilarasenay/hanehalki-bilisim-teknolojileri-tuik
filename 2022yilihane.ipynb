```python
import pandas as pd

# ======================================================
# GELİR GRUBU
# ======================================================

df = df.sort_values("hane gelir aylık").reset_index(drop=True)

df["gelir_grup"] = pd.qcut(
    df.index,
    q=5,
    labels=[1, 2, 3, 4, 5]
)

# ======================================================
# ONLINE SAĞLIK HİZMETİ KULLANIMI
# 1 = var, 2 = yok
# cinsiyet: 1 = erkek, 2 = kadın
# ======================================================

sonuc = []

for gelir in [1, 2, 3, 4, 5]:

    alt_df = df[df["gelir_grup"] == gelir]

    var_df = alt_df[alt_df["sağlık bilgisi arama"] == 1]
    yok_df = alt_df[alt_df["sağlık bilgisi arama"] == 2]

    var_cinsiyet_yuzde = var_df["cinsiyet"].value_counts(normalize=True) * 100
    yok_cinsiyet_yuzde = yok_df["cinsiyet"].value_counts(normalize=True) * 100

    sonuc.append({
        "yıl": 2021,
        "gelir": gelir,

        "var kadın": round(var_cinsiyet_yuzde.get(2, 0), 2),
        "var erkek": round(var_cinsiyet_yuzde.get(1, 0), 2),
        "var toplam": len(var_df),

        "yok kadın": round(yok_cinsiyet_yuzde.get(2, 0), 2),
        "yok erkek": round(yok_cinsiyet_yuzde.get(1, 0), 2),
        "yok toplam": len(yok_df)
    })

gelir_online_saglik_2021 = pd.DataFrame(sonuc)

# ======================================================
# EN ALT FREKANS SATIRI
# kadın/erkek/toplam hepsi frekans
# ======================================================

toplam_var = df[df["sağlık bilgisi arama"] == 1]
toplam_yok = df[df["sağlık bilgisi arama"] == 2]

toplam_var_cinsiyet = toplam_var["cinsiyet"].value_counts()
toplam_yok_cinsiyet = toplam_yok["cinsiyet"].value_counts()

toplam_satiri = pd.DataFrame([{
    "yıl": 2021,
    "gelir": "frekans",

    "var kadın": toplam_var_cinsiyet.get(2, 0),
    "var erkek": toplam_var_cinsiyet.get(1, 0),
    "var toplam": len(toplam_var),

    "yok kadın": toplam_yok_cinsiyet.get(2, 0),
    "yok erkek": toplam_yok_cinsiyet.get(1, 0),
    "yok toplam": len(toplam_yok)
}])

gelir_online_saglik_2021= pd.concat(
    [gelir_online_saglik_2021, toplam_satiri],
    ignore_index=True
)

display(gelir_online_saglik_2021)
```

```python
import pandas as pd
import numpy as np

# =========================================================
# YIL
# =========================================================
YIL = 2020

# =========================================================
# DEĞİŞKENLER
# =========================================================
gelir_col = "hanenin aylık toplam net geliri (20'li gelir grubu)"
randevu_col = "randevu alma (e-nabız,mhrs vb.)"

# =========================================================
# GELİR DEĞİŞKENİNİ NUMERİK YAP
# =========================================================
df["gelir_20li"] = pd.to_numeric(
    df[gelir_col],
    errors="coerce"
)

# =========================================================
# GELİRİ ZORLA 5 EŞİT GRUBA BÖL
# (HER GRUPTA YAKLAŞIK AYNI SAYIDA GÖZLEM)
# =========================================================
df["gelir_5li"] = pd.qcut(
    df["gelir_20li"].rank(method="first"),
    q=5,
    labels=[
        "1. grup",
        "2. grup",
        "3. grup",
        "4. grup",
        "5. grup"
    ]
)

# kontrol
print(df["gelir_5li"].value_counts().sort_index())


# =========================================================
# RANDEVU ALMA DEĞİŞKENİ
# =========================================================
df["randevu_alma"] = df[randevu_col].replace({
    "Evet": 1,
    "Hayır": 0,
    1: 1,
    2: 0
})


# =========================================================
# CROSS TABLO
# kadın/erkek yüzde
# toplam frekans
# =========================================================
sonuc = []

gelir_sirasi = [
    "1. grup",
    "2. grup",
    "3. grup",
    "4. grup",
    "5. grup"
]

for gelir in gelir_sirasi:

    alt = df[df["gelir_5li"] == gelir]

    # =====================================================
    # VAR
    # =====================================================
    var_df = alt[alt["randevu_alma"] == 1]

    var_toplam = len(var_df)

    var_kadin = round(
        ((var_df["cinsiyet"] == 2).sum() / var_toplam) * 100,
        2
    ) if var_toplam > 0 else 0

    var_erkek = round(
        ((var_df["cinsiyet"] == 1).sum() / var_toplam) * 100,
        2
    ) if var_toplam > 0 else 0


    # =====================================================
    # YOK
    # =====================================================
    yok_df = alt[alt["randevu_alma"] == 0]

    yok_toplam = len(yok_df)

    yok_kadin = round(
        ((yok_df["cinsiyet"] == 2).sum() / yok_toplam) * 100,
        2
    ) if yok_toplam > 0 else 0

    yok_erkek = round(
        ((yok_df["cinsiyet"] == 1).sum() / yok_toplam) * 100,
        2
    ) if yok_toplam > 0 else 0


    # =====================================================
    # SATIR EKLE
    # =====================================================
    sonuc.append({

        "yıllar": YIL,
        "gelir_grubu": gelir,

        "kadın_var": var_kadin,
        "erkek_var": var_erkek,
        "toplam_var": var_toplam,

        "kadın_yok": yok_kadin,
        "erkek_yok": yok_erkek,
        "toplam_yok": yok_toplam
    })


# =========================================================
# DATAFRAME
# =========================================================
final_tablo = pd.DataFrame(sonuc)


# =========================================================
# TOPLAM SATIRI
# =========================================================
toplam_var_df = df[df["randevu_alma"] == 1]
toplam_yok_df = df[df["randevu_alma"] == 0]

toplam_var = len(toplam_var_df)
toplam_yok = len(toplam_yok_df)

toplam_satiri = pd.DataFrame([{

    "yıllar": "",
    "gelir_grubu": "toplam",

    "kadın_var": round(
        ((toplam_var_df["cinsiyet"] == 2).sum() / toplam_var) * 100,
        2
    ),

    "erkek_var": round(
        ((toplam_var_df["cinsiyet"] == 1).sum() / toplam_var) * 100,
        2
    ),

    "toplam_var": toplam_var,

    "kadın_yok": round(
        ((toplam_yok_df["cinsiyet"] == 2).sum() / toplam_yok) * 100,
        2
    ),

    "erkek_yok": round(
        ((toplam_yok_df["cinsiyet"] == 1).sum() / toplam_yok) * 100,
        2
    ),

    "toplam_yok": toplam_yok

}])


# =========================================================
# FİNAL TABLO
# =========================================================
final_tablo = pd.concat(
    [final_tablo, toplam_satiri],
    ignore_index=True
)


# =========================================================
# GÖSTER
# =========================================================
display(final_tablo)


# =========================================================
# EXCEL AKTAR
# =========================================================
```

```python
import pandas as pd
import numpy as np

# =========================================================
# GELİR DEĞİŞKENİ
# =========================================================

gelir_col = "hanenin aylık toplam net geliri (20'li gelir grubu)"

# numerik yap
df["gelir_raw"] = pd.to_numeric(
    df[gelir_col],
    errors="coerce"
)

# =========================================================
# ZORLA EŞİT 5 GRUP
# =========================================================

df["gelir_5li_esit"] = pd.qcut(
    df["gelir_raw"].rank(method="first"),
    q=5,
    labels=[
        "1. grup",
        "2. grup",
        "3. grup",
        "4. grup",
        "5. grup"
    ]
)

# =========================================================
# FREKANS + YÜZDE
# =========================================================

gelir_sirasi = [
    "1. grup",
    "2. grup",
    "3. grup",
    "4. grup",
    "5. grup"
]

frekans = (
    df["gelir_5li_esit"]
    .value_counts()
    .reindex(gelir_sirasi)
)

yuzde = round(
    (
        df["gelir_5li_esit"]
        .value_counts(normalize=True)
        .reindex(gelir_sirasi)
    ) * 100,
    2
)

# =========================================================
# TABLO
# =========================================================

gelir_tablo = pd.DataFrame({
    "toplam": frekans,
    "yüzde": yuzde
})

# =========================================================
# TOPLAM SATIRI
# =========================================================

toplam_satiri = pd.DataFrame({
    "toplam": [gelir_tablo["toplam"].sum()],
    "yüzde": [100]
}, index=["toplam"])

gelir_tablo = pd.concat(
    [gelir_tablo, toplam_satiri]
)

display(gelir_tablo)

# =========================================================
# EXCEL AKTAR
# =========================================================
```
