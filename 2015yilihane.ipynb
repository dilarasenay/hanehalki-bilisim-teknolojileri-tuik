```python
from typing import Optional


def get_summary_text(result) -> Optional[str]:
    """result nesnesinden özet metni güvenli şekilde alır."""
    try:
        summary = result.summary()
        if summary is None:
            print("[UYARI] summary() None döndürdü.")
            return None

        text = summary.as_text()
        if not isinstance(text, str) or not text.strip():
            print("[UYARI] Özet metni boş veya geçersiz.")
            return None

        return text.strip()

    except AttributeError as e:
        print(f"[HATA] Geçersiz nesne yapısı: {e}")
        return None
    except Exception as e:
        print(f"[HATA] Beklenmeyen bir hata oluştu: {e}")
        return None


if summary_text := get_summary_text(result):
    print(summary_text)
```
